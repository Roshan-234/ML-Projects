```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Step 1: Load the data
data = pd.read_csv('nepse index.csv', encoding='latin1')
data['Close'] = data.rename(columns={'Close': 'NEPSE Index'}, inplace=True)
data.head()

```

```python
# # Step 2: Preprocess the data
# # Assuming the NEPSE index is in the 'NEPSE Index' column
# # Assuming the parameters are in columns: 'Close', 'High', 'Low', 'Absolute Change', 'Percentage Change', 'Turnover Values', 'Turnover Volume', 'Total Transaction'
features = ['High', 'Low', 'Absolute Change', 'Percentage Change', 'Turnover Values', 'Turnover Volume', 'Total Transaction']
target = 'NEPSE Index'
data['NEPSE Index'] = data['NEPSE Index'].str.replace(',', '').astype(float)
data['High'] = data['High'].str.replace(',', '').astype(float)
data['Low'] = data['Low'].str.replace(',', '').astype(float)
data['Turnover Values'] = data['Turnover Values'].str.replace(',', '').astype(float)
data['Turnover Volume'] = data['Turnover Volume'].str.replace(',', '').astype(float)
data['Total Transaction'] = data['Total Transaction'].str.replace(',', '').astype(float)#


```

```python
data['Percentage Change'] = data['Percentage Change'].str.replace(r'[^-0-9.]', '', regex=True).astype(float)
data['Absolute Change'] = data['Absolute Change'].str.replace(r'[^-0-9.]', '', regex=True).astype(float)
```

```python
# Step 3: Train-Test Split
train, test = train_test_split(data, test_size=0.2, shuffle=False)

```

```python
# Step 4: Neural Network Model
# Perform feature scaling
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train[features])
test_scaled = scaler.transform(test[features])

# Build the neural network model
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(train_scaled.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))  # Output layer

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(train_scaled, train['NEPSE Index'], epochs=1000, batch_size=32)

# Step 5: Make Predictions
nn_predictions = model.predict(test_scaled).flatten()

# Step 6: Evaluate Performance
nn_mse = mean_squared_error(test['NEPSE Index'], nn_predictions)

# Step 7: Print the MSE value
print(f"Neural Network MSE: {nn_mse}")
```
